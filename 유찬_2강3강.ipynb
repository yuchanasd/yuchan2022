```python
class AgeInfo:
  def up_age(self):
    self.age += 1
  def get_age(self):
    return self.age

fa = AgeInfo()
mo = AgeInfo()
me = AgeInfo()

fa.age = 50
mo.age = 49
me.age = 30

print("아빠의 나이:", fa.get_age())
print("엄마의 나이:", mo.get_age())
print("누나의 나이:", me.get_age())

print("1년 뒤...")
fa.up_age()
mo.up_age()
me.up_age()
print("아빠의 나이:", fa.get_age())
print("엄마의 나이:", mo.get_age())
print("누나의 나이:", me.get_age())
```

```python
import math
import matplotlib.pyplot as plt
class sin_graph:
  def generate_x_y(self):
    self.x = []
    self.y = []
    self.dx = (self.end-self.st)/(self.n-1)
    for i in range(self.n):
      x_= self.st + self.dx*i
      self.x.append(x_)
      self.y.append(math.sin(x_))
  def plot(self):
    plt.plot(self.x, self.y)
    plt.show()
g1 = sin_graph()
g1.st = 0
g1.end = 2*math.pi
g1.n = 10

g1.generate_x_y()
g1.plot()
```

```python
class sin_graph:
  def setting(self, st, ed, n):
    self.st = st
    self.ed = ed
    self.n = n
  def generate_x_y(self):
    self.x = []
    self.y = []
    self.dx = (self.ed-self.st)/(self.n-1)
    for i in range(self.n):
      x_= self.st + self.dx*i
      self.x.append(x_)
      self.y.append(math.sin(x_))
  def plot(self):
    plt.plot(self.x, self.y)
    plt.show()
g1 = sin_graph()
g1.setting(0, 2*math.pi, 100)
g1.generate_x_y()
g1.plot()
```

```python
g2 = sin_graph()
g2.setting(0, 4*math.pi, 100)
g2.generate_x_y()
g2.plot()
```
